```python
# we must some data preprocessing 
# we can start with missing values
hss.isna().sum()
```

```python
sns.distplot(hss['Radio']) # kind a normal distribution so Ä± can use mean, to fill missing values
```

```python
hss['Radio'].fillna(hss['Radio'].mean(),inplace=True)
```

```python
sns.distplot(hss['TV'])  # Ä± can use mean too 
```

```python
hss['TV'].fillna(hss['TV'].mean(),inplace=True)
```

```python
sns.distplot(hss['Social Media'])  # right skew data so Ä± can use median to filling the missing values because Ä± want to 
#declare to the average value
```

```python
hss['Social Media'].fillna(hss['Social Media'].median(),inplace=True)
```

```python
hss.isna().sum()
```

```python
sns.distplot(hss['Sales'])  # we can use mean kind a normal distribution 
```

```python
hss['Sales'].fillna(hss['Sales'].mean(),inplace=True)
```

```python
hss.isna().sum() # we have no missing values we got it
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
# feature engineering
X=hss[['TV','Radio','Social Media']]
y=hss['Sales']
# firstly Ä± want to add influencer because Ä± must encode .machine cannot get it with String data
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
# I should check it my dataset has distributed correctly ok everything is fine lets contuniue!
```

```python
lr=LinearRegression(normalize=True)
lr
```

```python
#We are fitting our values
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
# We must use Ä±nfluencer_encoded
```

```python
hss['Influencer_Encoded']=pd.factorize(hss['Influencer'])[0]
```

```python
# feature engineering
X=hss[['TV','Radio','Social Media']]
y=hss['Sales']
# firstly Ä± want to add influencer because Ä± must encode .machine cannot get it with String data
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
lr=LinearRegression(normalize=True)
lr
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
predictions=lr.predict(X_test)
score=mean_squared_error(predictions,y_test)
score
```

```python
# we can do hyperparameter tuning because our error is big
```

```python
lr_parameters={'n_jobs':[100,200,500],
    'fit_intercept':['auto','sqrt','log2'],
    'normalize':[6,8,10]
              }
```

```python
random_search=RandomizedSearchCV(lr,lr_parameters,cv=5)
random_search.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
# We have good values
# we can see best params nd bes accuracy
print(random_search.best_params_)
print(random_search.best_score_)
```

```python
predictions=random_search.predict(X_test)
score=mean_squared_error(predictions,y_test)
score
```
